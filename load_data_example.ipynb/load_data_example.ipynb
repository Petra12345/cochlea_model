# Heidelberg Spiking Datasets

The following code snipped illustrates how to open the Spiking HD Digits and/or Spiking Speech Comamnds datasets.
This example illustrates howto read the spiking dataset using www.pytables.org

```python
%matplotlib inline
```

```python
import tables
import numpy as np
```

```python
# The local directory where the dataset is stored

hdf5_file_path = "/home/zenkfrie/data/hdspikes/shd_train.h5"

fileh = tables.open_file(hdf5_file_path, mode='r')
units = fileh.root.spikes.units
times = fileh.root.spikes.times
labels = fileh.root.labels
```

```python
index = 0

print("Times (ms):", times[index])
print("Unit IDs:", units[index])
print("Label:", labels[index])
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(16,4))
idx = np.random.randint(len(times),size=3)
for i,k in enumerate(idx):
    ax = plt.subplot(1,3,i+1)
    ax.scatter(times[k],units[k], color="k", alpha=0.33, s=2)
    ax.set_title("Label %i"%labels[k])
    ax.axis("off")
```
